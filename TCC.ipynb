```python
html=pd.read_csv('C:/Users/danie/OneDrive/Documentos/Code Report/index.html', encoding='latin1')
x = str(html)
text = re.findall("<li>(.*?)<\/li>", x)
```

```python
text_file = open("C:/Users/danie/OneDrive/Documentos/Code Report/GmFonseca.txt", "w")
for val in text:
    text_file.write(val + '\n')
text_file.close()
```

```python
mismatched = 0
selector = 0
cannotResolveFile = 0
cannotResolveDirectory = 0
unterminatedStatement = 0
localVariable = 0
redundantCharacter = 0
xmlTag = 0
cannotResolveSymbol = 0
unresolvedPackage = 0
unresolvedClass = 0
noDirective = 0
unitOfMeasure = 0
unusedImport = 0
unusedParameter = 0
mismatchedParameters = 0
attemptToAssign = 0
typeNumber = 0
missingAssociated = 0
unusedField = 0
unusedReadonly = 0
unusedProperty = 0
unusedMethod = 0
unknownCSS = 0
unusedLocalVariable = 0
cannotResolveClass = 0
duplicateDeclaration = 0
propertyWidth = 0
propertyTransition = 0
missingRequired = 0
closingTag = 0
filepath = 'C:/Users/danie/OneDrive/Documentos/Code Report/WillList.txt'
with open(filepath) as fp:
   line = fp.readline()
   while line:
        if str(line).startswith('Mismatched property value'):
            mismatched = mismatched + 1
        if str(line).startswith('Selector'):
            selector = selector + 1
        if str(line).startswith('Cannot resolve file'):
            cannotResolveFile = cannotResolveFile + 1
        if str(line).startswith('Cannot resolve directory'):
            cannotResolveDirectory = cannotResolveDirectory + 1
        if str(line).startswith('Unterminated statement'):
            unterminatedStatement = unterminatedStatement + 1
        if str(line).startswith('Local variable'):
            localVariable = localVariable + 1
        if str(line).startswith('Redundant character escape'):
            redundantCharacter = redundantCharacter + 1
        if str(line).startswith('XML tag has empty body'):
            xmlTag = xmlTag + 1
        if str(line).startswith('Cannot resolve symbol'):
            cannotResolveSymbol = cannotResolveSymbol + 1
        if str(line).startswith('Unresolved package'):
            unresolvedPackage = unresolvedPackage + 1
        if str(line).startswith('Unresolved class'):
            unresolvedClass = unresolvedClass + 1
        if str(line).startswith('No directive is matched on attribute'):
            noDirective = noDirective + 1
        if str(line).startswith('Unit of measure'):
            unitOfMeasure = unitOfMeasure + 1
        if str(line).startswith('Unused import'):
            unusedImport = unusedImport + 1
        if str(line).startswith('Unused parameter'):
            unusedParameter = unusedParameter + 1
        if str(line).startswith('Mismatched parameters'):
            mismatchedParameters = mismatchedParameters + 1
        if str(line).startswith('Attempt to assign to const or readonly variable'):
            attemptToAssign = attemptToAssign + 1
        if str(line).startswith('Type number is not assignable to type string'):
            typeNumber = typeNumber + 1
        if str(line).startswith('Missing associated label'):
            missingAssociated = missingAssociated + 1
        if str(line).startswith('Unused field'):
            unusedField = unusedField + 1
        if str(line).startswith('Unused readonly field'):
            unusedReadonly = unusedReadonly + 1
        if str(line).startswith('Unused property'):
            unusedProperty = unusedProperty + 1
        if str(line).startswith('Unused method'):
            unusedMethod = unusedMethod + 1
        if str(line).startswith('Unknown CSS property'):
            unknownCSS = unknownCSS + 1
        if str(line).startswith('Unused local variable'):
            unusedLocalVariable = unusedLocalVariable + 1
        if str(line).startswith('Cannot resolve class'):
            cannotResolveClass = cannotResolveClass + 1
        if str(line).startswith('Duplicate declaration'):
            duplicateDeclaration = duplicateDeclaration + 1
        if str(line).startswith('Property width overwrites property width'):
            propertyWidth = propertyWidth + 1
        if str(line).startswith('Property transition is overwritten'):
            propertyTransition = propertyTransition + 1
        if str(line).startswith('Missing required \'alt\' attribute'):
            missingRequired = missingRequired + 1
        if str(line).startswith('Closing tag matches nothing'):
            closingTag = closingTag + 1
        line = fp.readline()
```
